```python
import pandas as pd
import numpy as np


from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('mydata.csv')
```

```python
data
```

```python
data.drop('institution', axis = 1, inplace = True)
```

```python
data.drop('year' ,axis = 1 , inplace = True)
```

```python
data
```

```python
data['broad_impact'].isnull().sum()
```

```python
data['broad_impact'] = data['broad_impact'].fillna(data['broad_impact'].mean())
```

```python
data['country'].value_counts()
```

```python
np.sum(data.isnull())
```

```python
encoder = LabelEncoder()
data['country'] = encoder.fit_transform(data['country'])
country_mapping = {index : label for index , label in enumerate(encoder.classes_)}
```

```python
country_mapping
```

```python
data
```

```python
y = data['world_rank']
X = data.drop('world_rank' , axis = 1)
```

```python
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X) , columns = X.columns)
```

```python
X
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y ,train_size = 0.8)
```

```python
model = LinearRegression()
model.fit(X_train , y_train)
```

```python
print(f"Linear Regression R^2: {model.score(X_test , y_test)}")
```
